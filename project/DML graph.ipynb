```python
import numpy as np
import matplotlib.pyplot as plt

# 固定参数
N = 100  # 数据点数
delta = 1.0  # 标准差

# 定义 DL 函数
def DL(NY, N, delta):
    term1 = 0.5 * NY * np.log2(N)
    term2 = 0.5 * NY * np.log2(2 * np.pi * delta**2 * np.e)
    return term1 + term2

# 模型自由参数数量范围（例如多项式阶数 + 1）
NY_values = np.arange(1, 20)
DL_values = [DL(ny, N, delta) for ny in NY_values]

# 绘图
plt.figure(figsize=(8, 5))
plt.plot(NY_values, DL_values, marker='o')
plt.title("Description Length DL vs. Number of Parameters $N_Y$")
plt.xlabel("Number of Free Parameters $N_Y$")
plt.ylabel("Description Length DL")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 生成一些假数据
N = 100
yi = np.random.normal(loc=5.0, scale=1.0, size=N)
y_pred = np.full(N, 5.0)  # 预测值为常数

# 定义似然函数
def likelihood(delta, y, y_pred):
    N = len(y)
    term1 = 1 / (np.sqrt(2 * np.pi) * delta)
    exponent = - ((y - y_pred) ** 2) / (2 * delta**2)
    return np.prod(term1 * np.exp(exponent))

# 在一系列 delta 值下计算似然
delta_values = np.linspace(0.1, 2.0, 100)
likelihood_values = [likelihood(d, yi, y_pred) for d in delta_values]

# 绘图
plt.figure(figsize=(8, 5))
plt.plot(delta_values, likelihood_values)
plt.title("Likelihood vs. δ")
plt.xlabel("δ")
plt.ylabel("Likelihood")
plt.grid(True)
plt.show()
```
