```python
import numpy as np
import matplotlib.pyplot as plt

# define function
def f(x):
    return (1 / (1 + np.exp(-x)) - 0.5) * 2

# Generates the range of values for x
x = np.linspace(-10, 10, 400)
y = f(x)

# plot
plt.plot(x, y, label=r'$f(x) = \left( \frac{1}{1 + e^{-x}} - 0.5 \right) \times 2$')
plt.title("Scaled Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# define sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# generate  x 
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

# plot
plt.plot(x, y, label=r'$f(x) = \frac{1}{1 + e^{-x}}$', color='blue')
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.show()
```
