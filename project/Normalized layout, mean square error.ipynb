```python
import numpy as np
import matplotlib.pyplot as plt

# 步骤1：原始归一化数据（你也可以换成自己的）
original = np.array([0.92, 0.92, 0.93, 0.93])

# 步骤2：减去均值（中心化）
mean_val = np.mean(original)
centered = original - mean_val

# 步骤3：映射回 (0, 1)
min_c = np.min(centered)
max_c = np.max(centered)
rescaled = (centered - min_c) / (max_c - min_c)

# 步骤4：输出处理前后数据
print("initial data:", np.round(original, 3))
print("Mean:", round(mean_val, 3))
print("After centralization:", np.round(centered, 3))
print("After remapping:", np.round(rescaled, 3))

# 步骤5：画图对比
plt.figure(figsize=(8, 4))
plt.plot(original, label="initial (0,1) data", marker='o')
plt.plot(rescaled, label="Re-centralize + map data", marker='x')
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.7, label='middle line 0.5')
plt.title("Further centralization after normalization")
plt.xlabel("sample number")
plt.ylabel("data")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 示例数据
data = np.array([0.46, 0.20, 0.63, 0.71])

# 均值
mean = np.mean(data)

# 均方差（即方差，Mean Squared Deviation）
variance = np.var(data)  # 默认为总体方差
# 或 np.var(data, ddof=1) 计算样本方差

# 标准差（standard deviation）
std_dev = np.sqrt(variance)

# 输出结果
print("数据：", data)
print("均值：", mean)
print("方差：", variance)
print("标准差：", std_dev)
```
